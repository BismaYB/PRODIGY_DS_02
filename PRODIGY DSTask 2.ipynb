```python
# Drop irrelevant columns
irrelevant_columns = ['PassengerId','Name', 'Ticket', 'Cabin','Embarked']
train_df = train_df.drop(irrelevant_columns, axis=1)
test_df = test_df.drop(irrelevant_columns, axis=1)
```

```python
train_df.isnull().sum()
```

```python
print(test_df.isnull().sum())
```

```python
# Verify data types
print(train_df.dtypes)
print(test_df.dtypes)
```

```python
# Handle missing values in Age column
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())

# Handle missing values in Fare column
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
```

```python
print(train_df.isnull().sum())
print(test_df.isnull().sum())
```

```python
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'], kde=False)
plt.title('Distribution of Age')
plt.show()
```

Age Distribution: The age distribution is skewed to the right, indicating that there are more younger passengers.

```python
# Distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Fare'], kde=False)
plt.title('Distribution of Fare')
plt.show()
```

Fare Distribution: The fare distribution is also skewed to the right, indicating that there are more passengers with lower fares.



```python
# Relationship between Age and Survived
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=train_df)
plt.title('Relationship between Age and Survived')
plt.show()
```

```python
# Relationship between Fare and Survived
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title('Relationship between Fare and Survived')
plt.show()
```

```python
# Relationship between Sex and Survived
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Relationship between Sex and Survived')
plt.show()
```

Sex-Based Survival: Female passengers have a significantly higher survival rate (74%) compared to male passengers (19%).



```python
# Relationship between Pclass and Survived
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Relationship between Pclass and Survived')
plt.show()
```

Class-Based Survival: Passengers in higher classes (1 and 2) have a significantly higher survival rate compared to those in lower classes (3).



```python
# 4. Combined Effect of Class and Gender on Survival
plt.figure(figsize=(12, 6))
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=train_df, kind='count', height=5, aspect=1.2, palette='husl')
plt.subplots_adjust(top=0.85)
plt.suptitle('Survival by Class and Gender', size=16)
plt.show()
```

```python

from sklearn.preprocessing import LabelEncoder

# Encode 'Sex' column into numeric values
label_encoder = LabelEncoder()
train_df['Sex'] = label_encoder.fit_transform(train_df['Sex'])  # male=1, female=0

# Compute correlation matrix with encoded data
correlation_matrix =train_df.corr()

# Display correlation matrix
print(correlation_matrix)
```

```python
# Visualize the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Relationships Between Variables:
Age and Survived: There is a negative correlation between age and survival rate, indicating that older passengers are less likely to survive.
Fare and Survived: There is a positive correlation between fare and survival rate, indicating that passengers with higher fares are more likely to
survive.
Sex and Survived: Female passengers have a significantly higher survival rate compared to male passengers.
Pclass and Survived: Passengers in higher classes have a significantly higher survival rate compared to those in lower classes.

Trends:
Increasing Survival Rate with Higher Fare: The survival rate increases with higher fares, indicating that passengers who paid more for their tickets
were more likely to survive.
Decreasing Survival Rate with Older Age: The survival rate decreases with older age, indicating that older passengers were less likely to survive.
Higher Survival Rate for Female Passengers: Female passengers have a significantly higher survival rate compared to male passengers.
Higher Survival Rate for Passengers in Higher Classes: Passengers in higher classes have a significantly higher survival rate compared to 
those in lower classes.
